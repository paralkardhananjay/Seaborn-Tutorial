```python
import seaborn as sns
import numpy as np
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width)
```

```python
sns.displot(iris['sepal_width'])
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip,data=tips,hue='smoker')
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip,data=tips,hue='tip',col='time',style='size', marker='s',color='blue')
```

```python
sns.displot(data=tips,kind='ecdf',x=tips.total_bill,col='time',hue='smoker',rug=True)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```
